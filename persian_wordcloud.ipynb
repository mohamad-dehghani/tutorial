```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from wordcloud_fa import WordCloudFa
```

```python
text = """
حادثه آکو یا انتقام چهل و هفت رونین یک رویداد تاریخی در قرن ۱۸ است که در آن یک گروه از رونین‌ها (سامورایی‌های بدون ارباب) انتقام مرگ ارباب خود را می‌گیرند. حادثه آکو به رشته حوادث به هم پیوسته‌ای گفته می‌شود که روز چهاردهم از ماه سوم سال چهاردهم گنروکو برابر با ۲۱ آوریل ۱۷۰۱ میلادی، با زخمی شدن کیرا یوشیناکا توسط آسانو ناگانوری آغاز شد و یک سال و نه ماه بعد، در روز چهاردهم ار ماه دوازده سال پانزدهم گنروکو برابر با ۳۰ ژانویه ۱۷۰۳ میلادی، با حمله رونین‌های آکو به عمارت کیرا پایان یافت. در روز چهاردهم مارس سال ۱۷۰۱ میلادی، توکوگاوا تسونایوشی، شوگون حاکم شوگون‌سالاری توکوگاوا، آسانو ناگانوری دایمیوی قلمروی آکو را مأمور پذیرش فرستاده‌ها از دربار امپراتوری در کیوتو قرار داد. شوگون همچنین یکی از اشراف بنام کیرا یوشیناکا را به عنوان مسئول آموزش تشریفات به آسانو ناگانوری منصوب کرد. روز آخر پذیرایی در قلعه ادو، آسانو شمشیر کوتاه واکیزاشی را کشید و قصد کشتن کیرا را داشت. آسانو به انجام سپوکو (خودکشی آئینی) محکوم شد، اما کیرا هیچ مجازاتی دریافت نکرد. شوگون‌سالاری (باکوفو) همچنین زمین‌های آسانو در قلمروی آکو را نیز مصادره کرد؛ بدین ترتیب سامورایی‌هایی که در خدمت وی بودند نیز کار و محل زندگی خود را از دست دادند و به رونین (سامورایی بدون ارباب) تبدیل شدند. یک سال و نه ماه بعد، در روز ۳۰ ژانویه سال ۱۷۰۳ میلادی، اُوایشی یُوشیئو، که یک سامورایی عالی‌رتبه در خدمت آسانو بود، با رهبری یک گروه چهل و هفت نفری از رونین‌ها به عمارت کیرا در ادو وارد شدند و کیرا را کشته و سپس سر او را بر روی قبر آسانو در معبد سنگاکو (واقع در توکیوی کنونی) گذاشتند. آنها سپس خود را به مقامات رسمی تحویل دادند و در ۲۰ ماه مارس ۱۷۰۳ میلادی، به حکم شوگون‌سالاری با انجام سپوکو به زندگی خود پایان دادند. 
"""
```

```python
wordcloud = WordCloudFa(persian_normalize=True,
                        background_color='white',
                        max_words=200,
                        no_reshape=True,
                        include_numbers=False,
                        collocations=False,
                        width=800,
                        height=400
                        )

wc = wordcloud.generate(text)
image = wc.to_image()
image.show()
image.save('wordcloud.png')
```
