ریشه‌یابی معنایی، فرآیند گروه‌بندی شکل‌های عطف یک کلمه است تا بتوان آن‌ها را به عنوان یک مورد تجزیه و تحلیل کرد که با ریشه‌ی معنایی یا فرم فرهنگ لغت کلمه شناسایی می‌شود.

<div id="sec_pos_tagger" style="direction:rtl;line-height:300%;">
 <font face="B Nazanin" size=6>
  <font face="B Titr" color=black size=5>
             برچسب‌گذار نقش کلمات(POS Tagger)
  </font>
  <p></p>
  <hr>
     برچسب‌گذاری نقش کلمات(Part-of-speech tagging)، فرایند برچسب‌گذاری یک واژه در یک متن است که آن برچسب، متناظر با نقش یا پاره‌گفتار(Part of speech) خاص آن واژه می‌باشد. برای مثال شناسایی واژه‌ها به عنوان «اسم»، «فعل»، «صفت»، «قید» و ... نوعی برچسب‌گذاری پاره‌های گفتار است.
با این حال برچسب‌گذاری واژه‌ها در یک متن می‌تواند مبنایی به جز پاره‌های گفتار داشته باشد.
     <br/>
     نوع دیگری از برچسب‌گذاری واژه‌ها با هدف شناسایی موجودیت‌های نامدار(Named Entity Recognition) انجام می‌شود که به برچسب‌گذاری IOB معروف است. در این نوع برچسب‌گذاری واژه اول موجودیت با B(Begin)، واژه‌های بعدی(در صورت وجود) با I(Inside)، و دیگر واژه‌ها با O(Outside) برچسب زده می‌شوند.
     <br/>
در ابزار Hazm، کلاس‌های POSTagger و IOBTagger به ترتیب برای برچسب‌گذاری نقش و موجودیت‌های نامدار استفاده می‌شوند. هر دو کلاس، از کلاس SequenceTagger که خود رابط(Interface) TaggerI ابزار NLTK را پیاده‌سازی می‌نماید، ارث‌بری می‌کنند.
       <br/>
     از متدهای tag و tag_sents می‌توان به ترتیب برای برچسب‌گذاری واژه‌های یک یا چند جمله استفاده کرد. به طور کلی برچسب‌گذارها نیاز به مدل‌های از پیش آماده شده دارند تا بر اساس آن بتوانند فرایند برچسب‌گذاری را انجام دهند. بنابراین در هنگام نمونه‌سازی از کلاس‌ها باید این مدل‌ها را به عنوان پارامتر در اختیار سازنده قرار داد.

 </font>
</div>

<div id="sec_chunker" style="direction:rtl;line-height:300%;">
 <font face="B Nazanin" size=6>
  <font face="B Titr" color=black size=5>
             بخش‌ساز(Chunker)
  </font>
  <p></p>
  <hr>
     بخش‌سازی(Chunking) فرایندی است که برای شناسایی نقش و عبارت‌های کوتاه موجود در یک جمله استفاده می‌شود. به عبارت دیگر می‌توان شناسایی گروه‌های زبانی ناهمپوشان (مانند عبارت‌های اسمی پایه) را هدف این فرایند قلمداد نمود.
        <br/>
     در ابزار hazm کلاس Chunker برای این هدف استفاده می‌شود. این کلاس رابط(Interface)ChunkParserI را پیاده‌سازی می‌نماید و از کلاس IOBTagger ارث‌بری می‌کند. از متدهای parse و parse_sents می‌توان به ترتیب برای بخش‌سازی یک یا چند جمله استفاده کرد.
        <br/>
     مشابه برچسب‌گذارها، بخش‌سازها نیز نیاز به مدل‌های از پیش آماده شده دارند تا بر اساس آن بتوانند فرایند بخش‌سازی را انجام دهند. همچنین بخش‌سازها برای انجام وظیفه‌شان نیاز به برچسب نقش کلمات دارند.
 </font>
</div>
